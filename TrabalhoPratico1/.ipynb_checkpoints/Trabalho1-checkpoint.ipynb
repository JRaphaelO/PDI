```python
def open_image(filename:str):
    img = Image.open(filename)
    pixels = np.asarray(img)
    
    img_width = (img.size[0])
    img_height = (img.size[1])
    return img, pixels, img_height, img_width
```

## Median e Sobel Filter

```python
def create_mean_mask(m, n):
    return np.array([[(1/int(m*n)) for i in range(m)] for l in range(n)])
```

```python
mean_mask = create_mean_mask(3, 3)

sobel_vertical = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_horizontal = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
```

```python
def inner_product(img, mask, isColeredMask=False):
    c = np.zeros(img.shape).astype(float)
    
    for i in range(len(img)):
        for j in range(len(img)):
            p = deepcopy(img[i, j])

            if not isColeredMask:
                c[i][j][0] = round(p[0] * mask[i][j])
                c[i][j][1] = round(p[1] * mask[i][j])
                c[i][j][2] = round(p[2] * mask[i][j])
            
            else:
                c[i][j][0] = round(p[0] * mask[i][j][0])
                c[i][j][1] = round(p[1] * mask[i][j][1])
                c[i][j][2] = round(p[2] * mask[i][j][2])
            
            
    
    r = np.sum(c[:, :, 0])
    g = np.sum(c[:, :, 1])
    b = np.sum(c[:, :, 2])
    
    if not isColeredMask:
        if r > 255: r = 255
        elif r < 0: r = 0

        if g > 255: g = 255
        elif g < 0: g = 0

        if b > 255: b = 255
        elif b < 0: b = 0
    
    return [r, g, b]
    
```

```python
def Correlation(pixels, height, width, mask, isColeredMask=False):
    img_pixels = deepcopy(pixels)
    
    m = len(mask[0])
    n = len(mask)
    
    max_height = (1 + (height - 1) - n - 1)
    max_width = (1 + (width - 1) - m - 1)
            
    i_dim = (len(pixels)  - n ) + 1
    j_dim = (len(pixels[0])  - m ) + 1
    
    # Cria uma imagem com as dimensoes da final
    img_conv = np.array([[[0, 0, 0] for j in range(j_dim)] for i in range(i_dim)])
    
    for i in tqdm(range(max_height)):
        for j in range(max_width):
            sub = img_pixels[i: i + n,j: j + m]
#             print('sub', sub, 'mask', mask)
            inner_result = inner_product(sub, mask, isColeredMask)
            img_conv[i][j] = inner_result
    
    return np.uint8(img_conv)    
```

```python
correlation_img = Correlation(pixels, img_height, img_width, mean_mask)
show_image(pixels, (60,6), "Original")
show_image(correlation_img, (60,6), "Mean Mask")
```

```python
correlation_img = Correlation(pixels, img_height, img_width, sobel_vertical)
show_image(pixels, (60,6), "Original")
show_image(correlation_img, (60,6), "Vertical Mask")
```

```python
correlation_img = Correlation(pixels, img_height, img_width, sobel_horizontal)
show_image(pixels, (60,6), "Original")
show_image(correlation_img, (60,6), "Horizontal Mask")
```

## Median

```python
def median_mask(img):
    return [np.median(img[:,:,0]), np.median(img[:,:,1]), np.median(img[:,:,2])]
```

```python
def median(pixels, height, width, m, n):
    max_height = ((height - 1) - n)
    max_width = ((width - 1) - m)
        
    i_dim = (len(pixels)  - n ) + 1
    j_dim = (len(pixels[0])  - m ) + 1
    
    # Cria uma imagem com as dimensoes da final
    img_conv = np.array([[[0, 0, 0] for j in range(j_dim)] for i in range(i_dim)])
    
    for i in range(max_height):
        for j in range(max_width):
            sub = pixels[i:i+n, j:j+m]
            inner_result = median_mask(sub)
            img_conv[i][j] = inner_result
    
    return np.uint8(img_conv)
```

```python
median_img = median(pixels, img_height, img_width, 7, 7)
show_image(pixels, (60, 5), "Original")
show_image(median_img, (60, 5), "Median Image")
```

```python
def get_point_correlation(img, pixels, mask, point_x, point_y, m, n, middle_point_x, middle_point_y):
    pixel_correlation = {
        "r_image": [],
        "r_mask": [],
        "g_image": [],
        "g_mask": [],
        "b_image": [],
        "b_mask": []
    }
    
    for x in range(m):
        for y in range(n):
            # Guarda todos os pixels das duas imagens que são envoltos pela máscara
            pixel_correlation['r_image'].append(pixels[point_x + (x - middle_point_x), point_y + (y - middle_point_y)][0])
            pixel_correlation['r_mask'].append(mask[x, y][0])
            
            pixel_correlation['g_image'].append(pixels[point_x + (x - middle_point_x), point_y + (y - middle_point_y)][1])
            pixel_correlation['g_mask'].append(mask[x, y][1])
            
            pixel_correlation['b_image'].append(pixels[point_x + (x - middle_point_x), point_y + (y - middle_point_y)][2])
            pixel_correlation['b_mask'].append(mask[x, y][2])
    
    # Faz a correlação para R, G, B. Levando em consideração os (R, G, B)'s das duas imagens para realizar as correlações
    # o ponto [0, 1] vai guardar o valor da correlação entre ambas as listas
    # a posteriori é feita a média das 3 correlações
    r_correlation = np.corrcoef(pixel_correlation['r_image'], pixel_correlation['r_mask'])[0, 1]
    g_correlation = np.corrcoef(pixel_correlation['g_image'], pixel_correlation['g_mask'])[0, 1]
    b_correlation = np.corrcoef(pixel_correlation['b_image'], pixel_correlation['b_mask'])[0, 1]
    
    correlation_mean = (r_correlation + g_correlation + b_correlation)/3.0
        
    return correlation_mean
```

```python
def get_correlation(img, pixels, mask, m, n):
    correlation = []
    
    middle_point_x = floor(m/2)
    middle_point_y = floor(n/2)
    
    # Atravessa a imagem sem extensão de zero, não vai passar dos limites da image
    # Para cada pixel da imagem vai guardar a média da correlação entre R, G, B
    for point_x in tqdm(range(middle_point_x, img.size[0] - middle_point_x)):
        for point_y in range(middle_point_y, img.size[1] - middle_point_y):
            correlation.append(
                {"value": get_point_correlation(img, pixels, mask, point_x, point_y, m, n, \
                                                    middle_point_x, middle_point_y),
                 "pixel": {
                     "x": point_x,
                     "y": point_y
                 }
                })

    return correlation
```

```python
# Pegar a largura e altura da imagem do olho
m = baboon_eye_img.size[0]
n = baboon_eye_img.size[1]

# Cronometrar o tempo da operação e pegar as correlações
start = time.time()
correlation = get_correlation(baboon_img, pixels_baboon, pixels_baboon_eye, m, n)
end = time.time()
```

```python
print(f'Elapsed Time: {str((end - start)/60).split(".")[0]}m{str(((end - start)/60)).split(".")[1][0:2]}s')
```

```python
# Plotar o gráfico com os picos da correlação
correlation_values = np.array([cor['value'] for cor in correlation])
peaks, _ = find_peaks(correlation_values, height=0.8)
correlation_pixels = [correlation[peak]['pixel'] for peak in peaks]
plt.plot(correlation_values)
plt.plot(peaks, correlation_values[peaks], "x")
plt.plot(np.zeros_like(correlation_values), "--", color="gray")
plt.show()
```

```python
def paint_image_peaks(img, pixels, correlation_pixels, m, n, color):
    # Para cada pixel dentro dos picos, pintar a borda da máscara ao redor dele
    for pixel in correlation_pixels:
        paint_mask_square(img, pixels, pixel, m, n, color)
```

```python
def paint_mask_square(img, pixels, pixel, m, n, color):
    
    middle_point_x = floor(m/2)
    middle_point_y = floor(n/2)
    
    # Iterar sobre as bordas da máscara e pintá-las  
    for x in range(m):
        pixels[pixel['x'] + (x - middle_point_x), pixel['y'] - middle_point_y] = color
        pixels[pixel['x'] - (x - middle_point_x), pixel['y'] + middle_point_y] = color
        
    for y in range(n):
        pixels[pixel['x'] - middle_point_x, pixel['y'] + (y - middle_point_y)] = color
        pixels[pixel['x'] + middle_point_x, pixel['y'] + (y - middle_point_y)] = color
```

```python
pixels_baboon = baboon_img.load()
pixels_baboon = paint_image_peaks(baboon_img, pixels_baboon, correlation_pixels, m, n, color=(255, 0, 255))
baboon_img.save('images/baboonWithPeaks.png')
displayImage(filename='images/baboonWithPeaks.png')
```
